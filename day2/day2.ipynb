### 1
What is the average click-through rate (CTR) for sponsored product ads for each product category that contains the substring 'Electronics' in its name during October 2024? This analysis will help determine which electronics-related categories are performing optimally.

```python
dim_product['product_category'].unique()
```

```python
dim_product
```

```python
dim_product_electronics = dim_product[dim_product['product_category'].str.contains('Electronics', na=False)]
dim_product_electronics
```

```python
oct24_ads = fct_ad_performance[(fct_ad_performance['recorded_date'].dt.month == 10) & (fct_ad_performance['recorded_date'].dt.year == 2024)]
oct24_ads
```

```python
df = pd.merge(left=oct24_ads, right=dim_product_electronics, on = 'product_id', how='inner')
df
```

```python
df['CTR'] = df['clicks'] / df['impressions']
```

```python
df
```

```python
avg_CTR = df['CTR'].mean()
print(f'{avg_CTR}')
```

```python
category_averages = df.groupby('product_category')['CTR'].mean()
print(category_averages)
```

### 2
Which product categories have a CTR greater than the aggregated overall average CTR for sponsored product ads during October 2024? This analysis will identify high-performing categories for further optimization. For this question, we want to calculate CTR for each ad, then get the average across ads by product category & overall.

```python
oct24_ads["CTR"] = oct24_ads['clicks'] / oct24_ads['impressions']
oct24_ads
```

```python
total_avg_CTR = oct24_ads['CTR'].mean()
total_avg_CTR
```

```python
combine_data = pd.merge(left = oct24_ads, right = dim_product, on = 'product_id', how = 'inner')
```

```python
total_category_averages = combine_data.groupby('product_category')['CTR'].mean()
total_category_averages
```

```python
high_avg_cat = total_category_averages[total_category_averages > total_avg_CTR]
high_avg_cat
```

```python
print(f'{high_avg_cat}')
```

### 3 
For the product categories identified in the previous question, what is the percentage difference between their CTR and the overall average CTR for October 2024? This analysis will quantify the performance gap to recommend specific categories for targeted advertising optimization.

```python
for category, avg_value in high_avg_cat.items():
    perc = (avg_value-total_avg_CTR) / total_avg_CTR * 100
    print(f'{category} : {perc:.2f}%')
```
